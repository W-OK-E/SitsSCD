```python
import os
import matplotlib.pyplot as plt
import rasterio

# Path to the folder containing images
folder = "/home/ubuntu/PS_10/sat_change/SitsSCD/outputs/running_visuals"

# Get sorted lists of gt and pred files
gt_files = sorted([f for f in os.listdir(folder) if f.startswith('gt') and f.endswith('.tif')])[-5:-1]
pred_files = sorted([f for f in os.listdir(folder) if f.startswith('pred') and f.endswith('.tif')])[-5:-1]

# Display images side by side
for gt_file, pred_file in zip(gt_files, pred_files):
    gt_path = os.path.join(folder, gt_file)
    pred_path = os.path.join(folder, pred_file)
    with rasterio.open(gt_path) as gt_src:
        gt_img = gt_src.read(1)

    with rasterio.open(pred_path) as pred_src:
        pred_img = pred_src.read(1)

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    axes[0].imshow(gt_img, cmap='gray')
    axes[0].set_title(f'GT: {gt_file}')
    axes[0].axis('off')

    axes[1].imshow(pred_img, cmap='gray')
    axes[1].set_title(f'Pred: {pred_file}')
    axes[1].axis('off')

    plt.show()
```
